```python
import numpy as np
import PIL.Image as Image
# crop image file to top 224*224 pixels
def crop_image(img):
    return img[:, :224, :224]

# read file from path with PIL and convert to numpy array
def read_image(path):
    img = Image.open(path).convert("RGB")
    img = np.array(img).transpose((2, 0, 1))  # to C, H, W
    return img

# crop and save to destination path
def process_and_save_image(src_path, dst_path):
    img = read_image(src_path)
    img_cropped = crop_image(img)
    img_cropped_pil = Image.fromarray(img_cropped.transpose((1, 2, 0)))  # to H, W, C
    img_cropped_pil.save(dst_path)
```

```python
import numpy as np
import PIL.Image as Image
def read_image(path):
    img = Image.open(path).convert('RGB')
    img = np.array(img).transpose((2, 0, 1))  # to C, H, W
    return img 
# unconditional shrink image to 224*224
def shrink_image(img):
    img_pil = Image.fromarray(img.transpose((1, 2, 0)))  # to H, W, C
    img_resized_pil = img_pil.resize((224, 224))
    img_resized = np.array(img_resized_pil).transpose((2, 0, 1))  # to C, H, W
    return img_resized
def shrink_and_save_image(src_path, dst_path):
    img = read_image(src_path)
    img_shrunk = shrink_image(img)
    img_shrunk_pil = Image.fromarray(img_shrunk.transpose((1, 2, 0)))  # to H, W, C
    img_shrunk_pil.save(dst_path)

# make directory "shrunk_images" if not exist
import os
os.makedirs("processed", exist_ok=True)
for i in range(1, 5):
    shrink_and_save_image(f"../test_files/test_{i}.jpg", f"../test_files/test_224_{i}.jpg")
```

```python
import numpy as np
from PIL import Image
# reduce resolution of image to 112*112; keep every second pixel
def reduce_image(img, sz):
    factor = 224//sz
    return img[:, ::factor, ::factor]
# read file from path with PIL and convert to numpy array
def read_image(path):
    img = Image.open(path).convert('RGB')
    img = np.array(img).transpose((2, 0, 1))  # to C, H, W
    return img 
# reduce and save image
def reduce_and_save_image(input_path, output_path, sz):
    img = read_image(input_path)
    img_reduced = reduce_image(img, sz)
    img_reduced_pil = Image.fromarray(img_reduced.transpose((1, 2, 0)))  # to H, W, C
    img_reduced_pil.save(output_path)

for i in range(1, 5):
    reduce_and_save_image(f"../test_files/test_224_{i}.jpg", f"../test_files/test_56_{i}.png", 56)
```

```python
#============================ Generate upscaled images with different algorithms ============================

import numpy as np
from PIL import Image
def read_image(path):
    img = Image.open(path).convert('RGB')
    img = np.array(img).transpose((2, 0, 1))  # to C, H, W
    return img 
# bilinear interpolation resize as baseline
def bilinear_resize(image, target_size):
    pil_image = Image.fromarray(image.transpose((1, 2, 0)))  # to H, W, C
    pil_resized = pil_image.resize((target_size, target_size), Image.BILINEAR)
    resized_image = np.array(pil_resized).transpose((2, 0, 1))  # to C, H, W
    return resized_image
def bicubic_resize(image, target_size):
    pil_image = Image.fromarray(image.transpose((1, 2, 0)))  # to H, W, C
    pil_resized = pil_image.resize((target_size, target_size), Image.BICUBIC)
    resized_image = np.array(pil_resized).transpose((2, 0, 1))  # to C, H, W
    return resized_image
def lanczos_resize(image, target_size):
    pil_image = Image.fromarray(image.transpose((1, 2, 0)))  # to H, W, C
    pil_resized = pil_image.resize((target_size, target_size), Image.LANCZOS)
    resized_image = np.array(pil_resized).transpose((2, 0, 1))  # to C, H, W
    return resized_image

input_image = read_image('../test_files/test_56.png')

# test bilinear resize
output_bilinear_image = bilinear_resize(input_image, 224)
output_image_pil = Image.fromarray(output_bilinear_image.transpose((1, 2, 0)))  # to H, W, C
output_image_pil.save('../test_files/outputs/test_bilinear_56.png')
# test bicubic resize
output_bicubic_image = bicubic_resize(input_image, 224)
output_bicubic_image_pil = Image.fromarray(output_bicubic_image.transpose((1, 2, 0)))  # to H, W, C
output_bicubic_image_pil.save('../test_files/outputs/test_bicubic_56.png')
# test lanczos resize
output_lanczos_image = lanczos_resize(input_image, 224)
output_lanczos_image_pil = Image.fromarray(output_lanczos_image.transpose((1, 2, 0)))  # to H, W, C
output_lanczos_image_pil.save('../test_files/outputs/test_lanczos_56.png')
```
